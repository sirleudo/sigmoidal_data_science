# Panorama do COVID-19 no Brasil

A COVID-19 é uma doença causada pelo **coronavírus**, também conhecida como **SARS-CoV-2**, doença essa que apresenta quadros clinicos de infecções assitomáticas a quadros graves, podendo levar o paciente ao óbito.

<center><img alt="Colaboratory logo" width="75%" src="https://github.com/sirleudo/sigmoidal_data_science/blob/master/Covid19.2.jpg?raw=true"></center>

A Organização Mundial de Saúde (OMS) afirma que em cerca de 80% dos casos detectados, o paciente que contraiu a **COVID-19** apenas são assintomáticos ou oligossintomáticos (poucos sintomas), já para os 20% restantes, serão necessários atendimento hospitalar, por esses apresentarem dificuldade em respirar e 5% podem necessitar de suporte ventilatório.

Com o intuíto de consciêntizar a população brasileira a respeito do **COVID-19**, irei realizar uma análise sobre a doença com os dados públicos coletados.

## Obtenção dos Dados

Os dados coletados para análise serão do portal [Our World in Data](https://ourworldindata.org/coronavirus).

Há uma grande dificuldade ao coletar dados de uma nova doença como o **COVID-19**, pois existem conflitos de informações nas esferas governamentais na divulgação dos dados.

### Dicionário de Variáveis
* `iso_code` - Códigos de país de três letras.
* `continent` - Continente da localização geográfica.
* `location` - Localização geográfica.
* `date` - Data de observação.
* `total_cases` - Total de casos confirmados de COVID-19.
* `new_cases` - Novos casos confirmados de COVID-19.
* `total_deaths` - Total de mortes atribuídas ao COVID-19.
* `new_deaths` - Novas mortes atribuídas ao COVID-19.
* `total_cases_per_million` - Total de casos confirmados de COVID-19 por 1.000.000 pessoas.
* `new_cases_per_million` - Novos casos confirmados de COVID-19 por 1.000.000 pessoas.
* `total_deaths_per_million` - Total de mortes atribuídas ao COVID-19 por 1.000.000 pessoas.
* `new_deaths_per_million` - Novas mortes atribuídas ao COVID-19 por 1.000.000 pessoas.
* `total_tests` - Total de testes para COVID-19.
* `new_tests` - Novos testes para COVID-19. 
* `new_tests_smoothed` - Novos testes para COVID-19 (7 dias suavizadas). 
* `total_tests_per_thousand` - Total de testes para COVID-19 por 1.000 pessoas.
* `new_tests_per_thousand` - Novos testes para COVID-19 por 1.000 pessoas.
* `new_tests_smoothed_per_thousand` - Novos testes para COVID-19 (7 dias suavizados) por 1.000 pessoas.
* `tests_units` - Unidades usadas pelo local para relatar seus dados de teste.
* `stringency_index` - Índice de Stringency de Resposta do Governo.
* `population` - População em 2020.
* `population_density` - Número de pessoas divididas por área terrestre.
* `median_age` - Idade mediana da população.
* `aged_65_older` - Parcela da população com 65 anos ou mais.
* `aged_70_older` - Parcela da população com 70 anos ou mais.
* `gdp_per_capita` - Produto interno bruto na paridade do poder aquisitivo.
* `extreme_poverty` - Parcela da população vivendo em extrema pobreza.
* `cvd_death_rate` - Taxa de mortalidade por doenças cardiovasculares.
* `diabetes_prevalence` - Prevalência de diabetes (% da população de 20 a 79 anos).
* `female_smokers` - Parte das mulheres que fumam.
* `male_smokers` - Parte dos homens que fumam.
* `handwashing_facilities` - Parcela da população com instalações básicas de lavagem de mãos no local.
* `hospital_beds_per_thousand` - Leitos hospitalares por 1.000 pessoas.
* `life_expectancy` - Expectativa de vida no nascimento.

## Análise Exploratória dos Dados

Para começar a analisar os dados do DataSet, vamos precisar importar alguns pacotes necessários para que em seguida possamos fazer a importação do DataSet para um DataFrame.

```
# importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style()

import warnings
warnings.filterwarnings('ignore')

# importar o dataset COVID-19 para um DataFrame
df = pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv');
```

### Análise Exploratória Inicial

Vamos visualizar as 05 (cinco) primeiras entradas do DataFrame para continuar nossa análise exploratória.

```
# visualizar as primeiras entradas
df.head()
```

Na sequência vamos verificar o tamanho do DataFrame, como o método `shape`, analisado quantas linhas e colunas possuem. Logo após, vamos imprimir os nomes de cada colunas desse Dataset.

```
# verificar o tamanho do DataFrame
print('Dimensões do Dataset','\nVariáveis: ', df.shape[1], "\n"
'Entrada: ', df.shape[0])
```

```
# extrair os nomes das colunas
df.columns
```

Quais os tipos de variávies encontradas nesse projeto?

```
# verificando o tipo de variáveis
df.dtypes
```

Quando se estar analisando um Dataset, um dos principais trabalho a se fazer é a verificação de **dados ausentes**, o qustionamento e\ou motivo desses dados estarem faltando no Dataset é um ponto fundamental para uma análise com mais precisão.

```
# verificar dados ausentes
round((df.isnull().sum() / df.shape[0]).sort_values(ascending=False),2)
```

Analisando os dados acima, verificamos que a coluna `data` não estar no formato de **datetime**, por esse motivo vamos transformar essa coluna em **datetime** para que possamos explorar o maior número de informações possíveis.

```
# transformar a coluna data em datetime
df.date = pd.to_datetime(df.date)
```

```
# verificar data rescente do DataFrame
df.date.max()
```

Paíse que tem mais número de casos confirmados segundo o DataFrame

```
# verificar o número de casos confirmados
df.loc[df.date == '2020-07-23'].sort_values(by="total_cases", ascending=False)
```

Constatamos que os países com mais número de casos confirmados são os seguintes:

* Estados Unidos - 3.970.906 casos
* Brasil - 2.227.514 casos
* Índia - 1.238.635 casos
* Russia - 789.190 casos

Agora, vamos verificar os casos de morte.

```
# verificar o número de mortes
df.loc[df.date == '2020-07-23'].sort_values(by="total_deaths", ascending=False)
```

O número de mortes confirmadas pela COVID-19 em 23 de Julho de 2020 era:

* Estados Unidos - 143.190
* Brasil - 82.771
* Grã-Bretanha - 45.501
* México - 41.190

Para melhorar a visualização, vamos plotar um gráfico de barras.

```
# gráfico de barras para total de mortes
total_deaths = df.loc[df.date == '2020-07-23', ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False)[1:6]

fig, ax = plt.subplots(figsize=(15,10))

ax.bar('location', 'total_deaths', data=total_deaths)

plt.tight_layout()
```

Para uma melhor análise da progressão da doença no mundo, vamos plotar um gráfico de linhas e análisar o número de mortes total.

```
# gráfico de linha para mortes no mundo
df.loc[df.location == "World", "total_deaths"].reset_index(drop=True).plot(figsize=(10,8));
```

### Análise Exploratória para o Brasil

Análise exploratória da doença no Brasil.

```
# Criar uma cópia do DataFrame apenas com o Brasil
df_brasil = df.loc[df.location == "Brazil"].copy()

df_brasil.head()
```

Data atual dos registro no Dataset.

```
df_brasil.date.max()
```

Desde o dia 31-12-2019 á 23-07-2020, vem sendo inseridos registros no Dataset. Identificaremos a data do primeiro registro de morte no Brasil.

```
# Identificar quando foi registrada a primeira morte no Brasil
df_brasil.loc[df_brasil['new_deaths'] != 0.0].head(1) 
```

```
# Identificar quantos dias demorou para registrarmos a primeira morte, considerando o primeiro caso de COVID-19 registrado.
# identificaçãso do primeiro caso
df_brasil.loc[df_brasil['new_cases'] != 0.0].head(1)
print('O primeiro caso foi registrado em {}'.format(df_brasil.date.loc[4153]))
```

```
# primeira morte
df_brasil.loc[df_brasil['new_deaths'] != 0.0].head(1)
print('O primeiro registrado de morte foi em {}'.format(df_brasil.date.loc[4174]))
```

Análise do primeiro caso confirmado até a primeira morte registrada.

```
# Identificar quantos dias demorou para registrarmos a primeira morte, considerando o primeiro caso de COVID-19 registrado.
# calculando tempo desde o primeiro caso até a primeira morte
from datetime import date
delta = df_brasil.date.loc[4174] - df_brasil.date.loc[4153]
print('Se passaram {} dias entre o primeiro caso e a primeira morte no Brasil.'.format(delta.days))
```

Finalizando a análise vamos plotar um gráficos de linha para compararmos a evolução dos casos e mortes no Brasil.

```
# Plotar gráficos de linha em escalas linear e logarítmica para os casos totais e mortes totais.
x = list(range(206))
y = df_brasil['total_cases']
y2 = df_brasil['total_deaths']

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(9, 7))

ax1.set_title('Casos no Brasil')
ax1.set_xlabel('Dias')
ax1.set_ylabel('Casos')
ax1.plot(x, y)

ax2.set_title('Casos no Brasil (Escala logarítmica)')
ax2.set_xlabel('Dias')
ax2.set_ylabel('Casos')
ax2.set_yscale('log')
ax2.plot(x, y)

ax3.set_title('Mortes no Brasil')
ax3.set_xlabel('Dias')
ax3.set_ylabel('Mortes')
ax3.plot(x, y2)

ax4.set_title('Mortes no Brasil (Escala logarítmica')
ax4.set_xlabel('Dias')
ax4.set_ylabel('Mortes')
ax4.set_yscale('log')
ax4.plot(x, y2)

plt.tight_layout();
```
