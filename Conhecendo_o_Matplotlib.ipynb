```
# importar pacotes
import numpy as np
import matplotlib.pyplot as plt
```

```
# gerar valores demonstrativos
np.random.seed(42)
x = np.arange(10)
y = np.random.normal(size=10)

# plotar valores
plt.plot(x, y)
plt.title('Exemplo')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.show()
```

```
fig, ax = plt.subplots()
```

```
x = np.arange(0, 100)

# plotar valores
fig, ax = plt.subplots()
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadrado')
ax.set_title('Abordagem OO')
ax.set_ylabel('Valores em Y')
ax.set_xlabel('Valores em X')
ax.legend()

fig.show()

```

```
# plotar 2 gr√°ficos
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8, 3))

# gr√°fico 1
ax[0].plot(x, x)
ax[0].set_title('Linear')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')

# gr√°fico 2
ax[1].plot(x, x**2)
ax[1].set_title('Quadrado')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')


fig.tight_layout();
```

```
x = np.arange(10)
plt.plot(x, y)
plt.plot(x, y**2)
plt.plot(x, y, label='linear')
plt.plot(x, y**2, label='quadrado')
plt.title('Interface Pyplot')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()

plt.show();
```
